```python
from sklearn.metrics import confusion_matrix
import numpy as np
import os
import pandas as pd
from sklearn.model_selection import train_test_split, StratifiedKFold
import matplotlib.pyplot as plt
from sklearn import metrics
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import RandomizedSearchCV
from sklearn.model_selection import cross_val_score
from sklearn.metrics import classification_report
from sklearn.ensemble import GradientBoostingClassifier
```

```python
data_cross = pd.read_csv("oasis_cross-sectional.csv")
data_long = pd.read_csv("oasis_longitudinal.csv")
print(data_cross.info())
print(data_long.info())
```

```python
print(data_cross.isna().sum())
print("\n")
data_long.isna().sum()
```

```python
data_cross.dropna(subset=['CDR'],inplace=True)
```

```python
data_cross.drop(columns=['ID','Delay'],inplace=True)
data_long = data_long.rename(columns={'EDUC':'Educ'})
data_long.drop(columns=['Subject ID','MRI ID','Group','Visit','MR Delay'],inplace=True)
```

```python
data = pd.concat([data_cross,data_long])
data.head(50)
```

```python
data.describe()
```

```python
# Select only numeric columns for correlation analysis
numeric_data = data.select_dtypes(include=['float64', 'int64'])
cor = numeric_data.corr()

# Plot the correlation matrix
plt.figure(figsize=(12, 9))
sns.heatmap(cor, xticklabels=cor.columns.values, yticklabels=cor.columns.values, annot=True)
plt.show()
```

```python
from sklearn.impute  import SimpleImputer
imputer = SimpleImputer ( missing_values = np.nan,strategy='most_frequent')

imputer.fit(data[['SES']])
data[['SES']] = imputer.fit_transform(data[['SES']])

# We perform it with the median
imputer = SimpleImputer ( missing_values = np.nan,strategy='median')

imputer.fit(data[['MMSE']])
data[['MMSE']] = imputer.fit_transform(data[['MMSE']])

```

```python
data
```

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
data['CDR'] = le.fit_transform(data['CDR'].values)
```

```python
data = pd.get_dummies(data)
```

```python
plt.bar(data['CDR'].unique(),data['CDR'].value_counts())
```

```python
data = data.drop(data[data['CDR']==3].index)
data.info()
```

```python
y = data.pop('CDR')
x = data
```

```python
x = StandardScaler().fit_transform(x)
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, random_state = 42,test_size=0.3)
```

```python
from xgboost import XGBClassifier
from sklearn.model_selection import RandomizedSearchCV

#... (your other imports and data preparation)

parametros_gb = {
    "learning_rate": [0.01, 0.025, 0.005, 0.1, 0.15, 0.2, 0.3, 0.8, 0.9],
    "max_depth": [3, 5, 8, 10, 15, 20, 25, 30, 40, 50],
    "subsample": [0.5, 0.618, 0.8, 0.85, 0.9, 0.95, 1.0],
    "n_estimators": range(1, 100)
}

NUM_FOLDS = 5  # Define the number of folds for cross-validation

model_gb = XGBClassifier()  # Use XGBClassifier from xgboost

gb_random = RandomizedSearchCV(estimator=model_gb, param_distributions=parametros_gb, n_iter=100, cv=NUM_FOLDS,
                               verbose=0, random_state=42, n_jobs=-1, scoring='accuracy')
gb_random.fit(X_train, y_train)

print("Best parameters:", gb_random.best_params_)
```

```python
model_gb = gb_random.best_estimator_
print("Test accuracy:", model_gb.score(X_test, y_test))
```

```python
# Cross-validation
cv_accuracy = cross_val_score(model_gb, x, y, cv=10, scoring='accuracy').mean()
print("Cross-validation accuracy:", cv_accuracy)
```

```python
# Use the best estimator from RandomizedSearchCV
best_gb_model = gb_random.best_estimator_

# Fit the model on the training data
best_gb_model.fit(X_train, y_train)

# Make predictions on the test data
y_pred = best_gb_model.predict(X_test)

# Now you can use the predictions for further analysis
cm = confusion_matrix(y_test, y_pred)
clr = classification_report(y_test, y_pred)

plt.figure(figsize=(8, 8))
sns.heatmap(cm, annot=True, vmin=0, fmt='g', cbar=False, cmap='Blues')
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix")
plt.show()

print("Classification Report:\n----------------------\n", clr)
```
